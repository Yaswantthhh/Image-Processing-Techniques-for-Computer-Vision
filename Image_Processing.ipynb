```python
import cv2
import numpy as np
from PIL import Image
from matplotlib import pyplot as plt

image_path = '/content/wallpaperflare.com_wallpaper (2).jpg'

image = cv2.imread(image_path)

resized_image = cv2.resize(image, (256, 256))

gray_image = cv2.cvtColor(resized_image, cv2.COLOR_BGR2GRAY)

blurred_image = cv2.GaussianBlur(gray_image, (5, 5), 0)

enhanced_image = cv2.equalizeHist(blurred_image)

_, binary_image = cv2.threshold(enhanced_image, 128, 255, cv2.THRESH_BINARY)

cv2.imwrite('/content/wallpaperflare.com_wallpaper (2).jpg', resized_image)
cv2.imwrite('/content/wallpaperflare.com_wallpaper (2).jpg', gray_image)
cv2.imwrite('/content/wallpaperflare.com_wallpaper (2).jpg', blurred_image)
cv2.imwrite('/content/wallpaperflare.com_wallpaper (2).jpg', enhanced_image)
cv2.imwrite('/content/wallpaperflare.com_wallpaper (2).jpg', binary_image)

print("Preprocessing complete. Images Saved")

images = [resized_image, gray_image, blurred_image, enhanced_image, binary_image]
titles = ['Resized Image', 'Gray Image', 'Blurred Image', 'Enhanced Image', 'Binary Image']

plt.figure(figsize=(15, 5))

for i in range(len(images)):
    plt.subplot(1, 5, i + 1)
    plt.imshow(images[i], cmap='gray'if len(images[i].shape) == 2 else None)
    plt.title(titles[i])
    # plt.axis('off')

# plt.tight_layout()
plt.show()
```

```python
import os
from PIL import Image
import numpy as np
import cv2
import concurrent.futures

image_folder= '/content/sample_data/augmented_images'
output_folder = '/content/resized_images'

os.makedirs(output_folder, exist_ok=True)

def process_image(filename):
  try:
    image_path = os.path.join(image_folder, filename)
    analog_image = Image.open(image_path)
    gray_image = analog_image.convert('L')
    gray_array = np.array(gray_image)
    _, binary_image = cv2.threshold(gray_array, 128, 255, cv2.THRESH_BINARY)

    output_path = os.path.join(output_folder, f"processed_{filename}")
    cv2.imwrite(output_path, binary_image)
    print(f"Processed and saved: {filename}")

  except Exception as e:
    print(f"Error processing {filename}: {e}")

with concurrent.futures.ThreadPoolExecutor() as executor:
  image_files = [f for f in os.listdir(image_folder) if f.endswith('.jpg') or f.endswith('.png')]
  executor.map(process_image, image_files)


```

```python
from tensorflow.keras.layers import BatchNormalization,Conv2D,MaxPooling2D
from tensorflow.keras.layers import Flatten,Dropout,Dense
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.preprocessing import image
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.regularizers import l2
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
import os
import cv2
from tensorflow.keras import optimizers
print(tf.__version__)
```

```python
image_size = 224
batch_size = 20

train = ImageDataGenerator(rescale=1/255)
test = ImageDataGenerator(rescale=1/255)

train_dataset = train.flow_from_directory('/content/Cars Dataset/train',
                                          target_size=(image_size, image_size),
                                          batch_size=batch_size,
                                          class_mode='categorical',
                                          shuffle=True)

test_dataset = test.flow_from_directory('/content/Cars Dataset/test',
                                        target_size=(image_size, image_size),
                                        batch_size=batch_size,
                                        class_mode = 'categorical',
                                        shuffle = False)


```

```python
print(train_dataset.class_indices)
validation_dataset = test_dataset
print(validation_dataset.class_indices)
print(validation_dataset.classes)
```

```python
image_size = 224
input_size = (image_size,image_size,3)
model = Sequential()
model.add(Conv2D(filters=32,kernel_size=(3,3),activation='relu',input_shape=input_size)) # Changed 'kernal_size' to 'kernel_size'
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Conv2D(filters=64,kernel_size=(3,3),activation='relu', input_shape=input_size)) # Changed 'kernal_size' to 'kernel_size'
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Conv2D(filters=64,kernel_size=(3,3),activation='relu', input_shape=input_size)) # Changed 'kernal_size' to 'kernel_size'
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Flatten())

model.add(Dense(units=128,activation='relu'))

model.add(Dropout(0.5))

model.add(Dense(units=7,activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss='categorical_crossentropy',
              optimizer=optimizers.RMSprop(learning_rate=2e-4),
              metrics=['acc'])
```

```python
history = model.fit(train_dataset,
                    epochs=10,
                    validation_data=test_dataset)
```

```python
acc = history.history['acc']
val_acc = history.history['val_acc']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs = range(len(acc))

plt.plot(epochs, acc, 'b', label='Training acc')
plt.plot(epochs, val_acc, 'r', label='Validation acc')
plt.title('Training and validation accuracy')
plt.legend()

plt.figure()

plt.plot(epochs, loss, 'b', label='Training loss')
plt.plot(epochs, val_loss, 'r', label='Validation loss')
plt.title('Training and validation loss')
plt.legend()

plt.show()
```
